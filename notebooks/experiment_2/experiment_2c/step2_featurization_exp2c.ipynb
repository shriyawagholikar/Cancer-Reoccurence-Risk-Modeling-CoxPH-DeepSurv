```python
# define the main train and test dataset path
MAIN_TRAIN_DATA_PATH = "/content/drive/MyDrive/1:1_Shriya_Wagholikar/Datasets/train_data.csv"
MAIN_TEST_DATA_PATH = "/content/drive/MyDrive/1:1_Shriya_Wagholikar/Datasets/test_data.csv"
```

```python
import pandas as pd
import numpy as np
# read the main train data
train_df = pd.read_csv(MAIN_TRAIN_DATA_PATH)
train_df.head()
```

```python
# read the main test data
test_df = pd.read_csv(MAIN_TEST_DATA_PATH)
test_df.head()
```

Select 100 best features using Selectkbest

```python
from sklearn.feature_selection import SelectKBest, f_classif
# define the components
COMPONENTS = 100
# select x and y values
X = train_df.drop(columns=['Disease Free (Months)', 'DFS_STATUS_ENCODED'])
y = train_df['DFS_STATUS_ENCODED']
columns = np.array(list(X.columns))
# Select the top 50 features
selector = SelectKBest(f_classif, k=COMPONENTS)
X.columns = X.columns.astype(str)
X_new = selector.fit_transform(X, y)
# filter the dataframe with the selected features
selected_features = list(map(str, columns[selector.get_support()])) + ['Disease Free (Months)', 'DFS_STATUS_ENCODED']
print(selected_features)
# select training and testing data based on the filter
train_df_filter = train_df[selected_features]
test_df_filter = test_df[selected_features]
```

Select Top 50 Features Using SFS (Sequential Feature Selector)

```python
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
# define the components
COMPONENTS = 50
# Estimator
model = LogisticRegression()
# select x and y values
X_ = train_df_filter.drop(columns=['Disease Free (Months)', 'DFS_STATUS_ENCODED'])
y_ = train_df_filter['DFS_STATUS_ENCODED']
# Define the scoring metric as it determines how the performance of a feature is evaluated during the feature selection process
scoring = 'f1'  # Use 'f1' for binary classification. For multiclass, use 'f1_macro' or 'f1_weighted'
# Apply t-SNE
selector = SequentialFeatureSelector(model, n_features_to_select = COMPONENTS,scoring =scoring ,n_jobs = -1)


# Select the top 50 features
X_.columns = X_.columns.astype(str)
X_new = selector.fit_transform(X_, y_)
columns = np.array(list(X_.columns))
# filter the dataframe with the selected features
selected_features_sfs = list(map(str, columns[selector.get_support()])) + ['Disease Free (Months)', 'DFS_STATUS_ENCODED']
print(selected_features_sfs)
# select training and testing data based on the filter
train_df_final = train_df_filter[selected_features_sfs]
test_df_final = test_df_filter[selected_features_sfs]
```

```python
# define the train and test path
TRAIN_PATH = "/content/drive/MyDrive/1:1_Shriya_Wagholikar/Datasets/experiment_2c/exp2c_sk_sfs_train.csv"
TEST_PATH = "/content/drive/MyDrive/1:1_Shriya_Wagholikar/Datasets/experiment_2c/exp2c_sk_sfs_test.csv"
# save the train csv
train_df_final.reset_index(drop = True, inplace = True)
train_df_final.to_csv(TRAIN_PATH, index = False)
# save the test csv
test_df_final.reset_index(drop = True, inplace = True)
test_df_final.to_csv(TEST_PATH, index = False)
```
