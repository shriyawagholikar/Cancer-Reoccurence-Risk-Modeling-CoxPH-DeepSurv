```python
import pandas as pd
import numpy as np

train_df = pd.read_csv(MRNA_TRAIN_DATA_PATH)
train_df.head()
```

```python
test_df = pd.read_csv(MRNA_TEST_DATA_PATH)
test_df.head()
```

Select 50 Features

```python
from sklearn.feature_selection import SelectKBest, f_classif

COMPONENTS = 50
# select x and y values
X = train_df.drop(columns=['Disease Free (Months)', 'DFS_STATUS_ENCODED']) #these need to be predicted
y = train_df['DFS_STATUS_ENCODED']
# select top 50 features
selector = SelectKBest(f_classif, k=COMPONENTS)
X.columns = X.columns.astype(str)
X_new = selector.fit_transform(X,y) #will train the model and transform the data both

selected_features = list(X.columns[selector.get_support()]) + ['Disease Free (Months)', 'DFS_STATUS_ENCODED']
print(selected_features)

# select training and testing data based on the filter
train_df_filter = train_df[selected_features]
test_df_filter = test_df[selected_features]
```

Identify Highly Co-Related Features and Drop

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
VIF_THRESHOLD = 5
#function to calculate VIF
def calculate_vif(df):
    vif = pd.DataFrame()
    vif["Feature"] = df.columns
    vif["VIF"] = [variance_inflation_factor(df.values, i) for i in range(df.shape[1])]
    return vif

vif = calculate_vif(train_df_filter.drop(columns=['Disease Free (Months)', 'DFS_STATUS_ENCODED'])) #call function and calculate vif
#remove features with vif greater than threshold value (50%)
high_vif_features = vif[vif['VIF'] > VIF_THRESHOLD]['Feature']
train_df_selected = train_df_filter.drop(columns=high_vif_features)
test_df_selected = test_df_filter.drop(columns=high_vif_features)
```

Dropping the Columns Based on Variance

```python
events = train_df_selected['DFS_STATUS_ENCODED'].astype(bool)
columns_to_drop = []
for column in train_df_selected.columns:
    if column not in ['DFS_STATUS_ENCODED', 'Disease Free (Months)']:
        variance = train_df_selected[column].var()
        var_positive = train_df_selected.loc[events, column].var()
        var_negative = train_df_selected.loc[~events, column].var()
        # Check if variance is lower than 0.5 in either case
        if var_positive < 0.5 or var_negative < 0.5:
            columns_to_drop.append(column)
print("Cols to drop: {}".format(columns_to_drop))
# Drop the identified columns from the train and test set
train_df_final = train_df_selected.drop(columns=columns_to_drop)
test_df_final = test_df_selected.drop(columns=columns_to_drop)
```

```python
TRAIN_PATH = "/content/drive/MyDrive/1:1_Shriya_Wagholikar/Datasets/experiment_1/exp1_selectKBest_vif_train.csv"
TEST_PATH = "/content/drive/MyDrive/1:1_Shriya_Wagholikar/Datasets/experiment_1/exp1_selectKBest_vif_test.csv"

train_df_final.reset_index(drop = True, inplace = True)
train_df_final.to_csv(TRAIN_PATH, index = False)
test_df_final.reset_index(drop = True, inplace = True)
test_df_final.to_csv(TEST_PATH, index = False)
```
